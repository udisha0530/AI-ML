```python
#part 1 answer
emoji_list = [':slightly_smiling_face:', ':pensive_face:', ':disguised_face:',
              ':headphone:', ':astonished_face:', ':hamburger:', ':grinning_face_with_smiling_eyes:',
              ':panda:', ':fearful_face:', ':loudly_crying_face:']

emoji_dict = {}

for emoji in emoji_list:
    user_input = input(f"Give 5 words that describe the emoji {emoji}: ")
    words = set([word.strip().lower() for word in user_input.split(';')])
    emoji_dict[emoji] = words

print(emoji_dict)
```

```python
#part2 answer
import pickle
file_path = 'emoji_data.pkl'
with open(file_path, 'wb') as file:
    pickle.dump(emoji_dict, file)
```

```python
#part3 answer
file_path = 'emoji_data.pkl'
with open(file_path, 'rb') as file:
    emoji_dict = pickle.load(file)
emoji_dict.items()    
```

```python
#part4 answer
sentence = input("Write a sentence containing some of the words used to describe the emojis: ")
for code, words in emoji_dict.items():
    for w in words:
        if w in sentence:
            sentence = sentence.replace(w, f"{w} {emoji.emojize(code)}")
print(sentence)







```
